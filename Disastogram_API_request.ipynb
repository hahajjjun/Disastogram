## Disastogram_text_data_api_request

```python
import urllib.request
from xml.etree import ElementTree as ET
import numpy as np  # for using pandas
import pandas as pd  # for using dataframes
```

```python
xmlDict = {}
for i in range(1,1000):
    requestURL = "http://apis.data.go.kr/1741000/DisasterMsg3/getDisasterMsg1List?ServiceKey=2cfRC75KcC8azzCGSwG44LEWXOwn4ClJJeLuqenrp5moIYttbwqmHMfCatGJi5SJLoQupmkrsLfImwS3PyM5og%3D%3D&type=xml&pageNo="+str(i)+"&numOfRows=500"
    root = ET.parse(urllib.request.urlopen(requestURL)).getroot()
    flag = 0
    for sitemap in root:
        if flag !=0:
            children = sitemap.getchildren()
            if "행정안전부" not in children[4].text and "경찰청" not in children[4].text and "공사" not in children[4].text and "국민안전처" not in children[4].text:
                xmlDict[children[2].text] = children[4].text
        flag = 1
    print(len(xmlDict))
```
